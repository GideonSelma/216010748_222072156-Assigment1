```julia
using CairoMakie, StatsBase

# Count accelerations
accel_counts = countmap(fetal_health_df.accelerations)

# Sort by acceleration value
accel_vals = collect(keys(accel_counts))
counts = collect(values(accel_counts))
sorted_indices = sortperm(accel_vals)
accel_vals = accel_vals[sorted_indices]
counts = counts[sorted_indices]

# Generate color vector
n = length(accel_vals)
colors = [get(cgrad(:RdPu), i / n) for i in 1:n]

# Plot
fig = Figure()
ax = Axis(fig[1, 1], title = "Accelerations", xlabel = "Acceleration", ylabel = "Count")

barplot!(ax, accel_vals, counts; color = colors)

fig
```

```julia
using DataFrames, Statistics, CairoMakie

# Select only the numeric columns
numeric_df = select(fetal_health_df, names(fetal_health_df, eltype.(eachcol(fetal_health_df)) .== Float64))

# Computing the correlation matrix
corr_matrix = cor(Matrix(numeric_df))
labels = names(numeric_df)
n = length(labels)

# Creating heatmap
fig = Figure(size = (800, 600))
ax = Axis(fig[1, 1], title = "Correlation Heatmap", xticklabelrotation = π/3,
          xticks = (1:n, labels), yticks = (1:n, labels))

heatmap!(ax, corr_matrix; colormap = :RdPu, colorrange = (-1, 1))

Colorbar(fig[1, 2], limits = (-1, 1), colormap = :RdPu)

fig
```
